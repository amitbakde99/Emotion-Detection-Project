```python
import cv2
import matplotlib.pyplot as plt
#pip install deepface
from deepface import DeepFace
```

```python
img= cv2.imread("/kaggle/input/fer2013/train/happy/Training_10066226.jpg")
```

```python
plt.imshow(img)
```

```python
predictions = DeepFace.analyze(img, enforce_detection=False)
print("="*50)
print("Gender :", predictions['gender'])
print("Emotion :", predictions['dominant_emotion'])
print("Race :", predictions['dominant_race'])
print('Age :', predictions['age'])
print("="*50)
```
